```python
# create a plot for erf

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def normcdf(mu, sigma, x):
    return norm.cdf((x+mu)/sigma)

x = np.linspace(-10, 10, 1000)
X_1 = normcdf(0,1,x)
X_2 = normcdf(0,2,x)
X_3 = normcdf(0,3,x)
X_4 = normcdf(0,4,x)

plt.plot(x, X_1, label="N(0,1)")
plt.plot(x, X_2, label="N(0,2)")
plt.plot(x, X_3, label="N(0,3)")
plt.plot(x, X_4, label="N(0,4)")
plt.legend()
plt.xlabel('x')
plt.ylabel('$\Phi(x)$')
plt.yticks([0, 1/2 ,1])
plt.grid(True)
plt.savefig('phi.svg')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x= np.linspace(-10, 10, 1000)

X_1 = x
X_1 = norm(0,1).pdf(x)
f = np.sqrt(np.pi)/np.pi * np.exp(-X_1**2 + 2*X_1 -1)

# plot
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('$f_X1(x_1 | x_2 = 2)$')
plt.grid(True)
plt.show()
```
