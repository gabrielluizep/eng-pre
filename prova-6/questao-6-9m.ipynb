```octave
N = 10000;
Nt = 1000;
t = linspace(-3, 3, Nt);

figure; hold on; grid on;
for A = [-1/2, 1/2]
    for T = [-1/2, 1/2]
        X = A * (abs(t - T) < 0.5);
        plot(t, X, 'DisplayName', sprintf('A=%.2f, T=%.2f', A, T), 'LineWidth', 4);
        hold on;
    end
end

legend;
ylim([-1.5 1.5])
```

```octave
X = zeros(N, Nt);
for i = 1 : N
  A = (randi([0, 1]) * 2 - 1) / 2;
  T = (randi([0, 1]) * 2 - 1) / 2;
  
  X(i, :) = A * (abs(t - T) < 0.5) ;
end

muX_sim = mean(X);
muX_teo = zeros(Nt);

figure; grid on; hold on;
plot(t, muX_sim, 'y', 'LineWidth', 4);
plot(t, muX_teo, 'b--', 'LineWidth', 2);
title('Função média'); xlabel('t'); ylabel('\mu_X(t)');
ylim([-1 1])
```

```octave
t1 = t; t2 = t;
[T1, T2] = meshgrid(t1, t2);
CX_sim = cov(X, X);
CX_teo = 1/4 .* (1/2 .* (abs(T1 + 1/2) < 0.5) .* (abs(T2 + 1/2) < 0.5) + 1/2 .* (abs(T1 - 1/2) < 0.5) .* (abs(T2 - 1/2) < 0.5));

figure;
subplot(1, 2, 1); grid on; hold on; view(45, 45);
surf(t1, t2, CX_sim);
shading flat;
title('C[X]_{sim}'); xlabel('t_1'); ylabel('t_2'); zlabel('C_X(t_1, t_2)');
zlim([0 0.25])

subplot(1, 2, 2); grid on; hold on; view(45, 45);
surf(t1, t2, CX_teo);
shading flat;
title('C[X]_{teo}'); xlabel('t_1'); ylabel('t_2'); zlabel('C_X(t_1, t_2)');
zlim([0 0.25])
```
